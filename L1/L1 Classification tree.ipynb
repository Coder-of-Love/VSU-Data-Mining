```python
import numpy as np
import pandas as pd
```

```python
fname = "велосипед1(7).xls"
data = pd.read_excel(fname)
data.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for i in range(data.shape[1]):
    data.iloc[:, i] = le.fit_transform(data.iloc[:, i])

X = data.iloc[:,:-1]
Y = data['Приобрел велосипед']

lim = round(X.shape[0]*0.7)
X_train = X[:lim]
X_test = X[lim:]
```

```python
Y = data['Приобрел велосипед']
Y_train = Y[:lim]
Y_test = Y[lim:]
Y.head()
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, Y_train)
```

```python
Y_pred = clf.predict(X_test)
Y_pred.dtype
```

```python
Y_pred = clf.predict(X_test)
accuracy = 0

for i in range(Y_pred.shape[0]):
    if (Y_pred[i] == Y_test.data[i]):
        accuracy += 1
    print("Предсказано:", Y_pred[i], "Правильно:", Y_test.data[i])

print("Точность:", accuracy/Y_pred.shape[0]*100, '%')
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
#graph.render("iris")
dot_data = tree.export_graphviz(clf, out_file=None, 
                                feature_names=data.iloc[0,1:],  
                                class_names=['Купил(а) велосипед', 'Не купил(а) велосипед'],  
                                filled=True, rounded=True,  
                                special_characters=True)  
graph = graphviz.Source(dot_data)  
graph.render("Велосипед")
```
